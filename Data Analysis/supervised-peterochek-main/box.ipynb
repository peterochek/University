```python
import pandas as pd
```

```python
box = pd.read_csv('./boxoffice.csv')
```

```python
box.head()
```

```python
box['release_date'] = box['release_date'].str.split('(').str[0]  # Remove the part inside parentheses
box['release_date'] = pd.to_datetime(box['release_date'], errors='coerce')  # Convert to datetime format
box['year'] = box['release_date'].dt.year.astype('Int64')
box['month'] = box['release_date'].dt.month.astype('Int64')
box['day'] = box['release_date'].dt.day.astype('Int64')

# Remove the original "release_date" column
box.drop(columns='release_date', inplace=True)

# Convert the percentage columns to float values
percent_columns = ['domestic_percent', 'international_percent']
for col in percent_columns:
    box[col] = box[col].str.rstrip('%').replace('â€“', '0').astype(float) / 100

# Convert the dollar columns to int columns
dollar_columns = ['domestic', 'international', 'worldwide', 'domestic_opening', 'budget']
box[dollar_columns] = box[dollar_columns].replace('[\$,]', '', regex=True)
for col in dollar_columns:
    box[col] = pd.to_numeric(box[col], errors='coerce')
    box[col] /= 1_000_000
    
box = box.drop(columns=['title'])
```
