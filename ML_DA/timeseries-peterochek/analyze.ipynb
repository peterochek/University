```python
from timeseries.data import Data

data_class = Data()

data = data_class.raw_data()
```

```python
data.head()
```

```python
data.plot()
```

```python
data = data.values
```

```python
from statsmodels.tsa.stattools import adfuller as adf

adf_value, p_value, lag, nobs, critical, info = adf(data)
print(f'{adf_value=}, {p_value=}, {lag=}')
print(f'{nobs=}, {critical=}, {info=}')
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

import matplotlib.pyplot as plt

plot_pacf(data)
plot_acf(data)
plt.show()
```

```python
import numpy as np


data = np.log(data)
```

```python
plot_pacf(data)
plot_acf(data)
```

```python
plot_pacf(np.diff(data))
plot_acf(np.diff(data))
plt.show()
```

###  

```python
plot_pacf(np.diff(data, 12))
plot_acf(np.diff(data, 12))
plt.show()
```

```python
plot_pacf(np.diff(data, 4))
plot_acf(np.diff(data, 4))
plt.show()
```
