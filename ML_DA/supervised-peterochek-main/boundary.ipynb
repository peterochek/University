```python
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
from sklearn import datasets


X, y = datasets.load_iris(return_X_y=True)
```

```python
def plot_decision_boundary(clf):
    print("F1 = ", cross_val_score(clf, X, y, cv=5, scoring='f1_macro').mean())

    X2d = PCA(n_components=2).fit_transform(X) 
    X2d = X[:,0:2]
    
    clf.fit(X2d, y)
    disp = DecisionBoundaryDisplay.from_estimator(clf, X2d, response_method="predict", alpha=0.5)
    disp.ax_.scatter(X2d[:, 0], X2d[:, 1], c=y, edgecolor="k")
    plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier

plot_decision_boundary(RandomForestClassifier())
```
